```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
import statistics as stat
import scipy.stats as st
import random as rand

que = 'SELECT TOP 20000 gaia.source_id, gaia.ra, gaia.dec, gaia.pmra, gaia.pmdec, \
    gaia.parallax, gaia.phot_g_mean_mag,gaia.phot_bp_mean_mag,gaia.phot_rp_mean_mag \
    FROM gaiadr2.gaia_source as gaia WHERE gaia.ra >= 278.26522 and gaia.ra <= 279.92663 and \
    gaia.dec <= -23.15043 and gaia.dec >= -24.67088 and gaia.parallax > 0.101 and \
    gaia.parallax < 0.104 and gaia.phot_g_mean_mag is not null and gaia.phot_bp_mean_mag is not null \
    and gaia.phot_rp_mean_mag is not null and gaia.pmra is not null and gaia.pmdec is not null'

def pc_ly(x):
    return (1/(float(x)/1000))*3.26156
def mas_deg(x):
    return x*(1/1000)/3600
def abs_mag(x,y):
    return x - (5 * (np.log10(y/10))) 
def color(x,y):
    return [y[i]-x[i] for i in range(len(x))]
def z_score(x,y,z):
    return (x-y)/z

ngc = pd.read_csv('/Users/landonbuechner/Documents/Git Hub Repositories/Hyades-Open-Cluster/ngc.csv')

ra0, dec0 = ngc['ra'], ngc['dec'] 
pmra, pmdec = mas_deg(ngc['pmra']), mas_deg(ngc['pmdec'])

plt.figure(figsize=(8,8))
plt.style.use('seaborn')
plt.xlim([280,278.2])
plt.ylim([-23.1,-24.7])
plt.xlabel('ra')
plt.ylabel('dec')
plt.title('Initial Data - {} stars'.format(len(ngc)))
for i in range(len(ngc)):
    ra = []
    dec = []
    for x in range(1,5000,10):
        ra.append(ra0[i]+(pmra[i]*x*2))
        dec.append(dec0[i]+(pmdec[i]*x*2))
    plt.scatter(ra, dec, s = .2, c = 'k'  )
plt.show()
ngc.head()
```

```python
M_g = [abs_mag(ngc['phot_g_mean_mag'][i],ngc['parallax'][i]) for i in range(len(ngc['ra']))]
M_rp = [abs_mag(ngc['phot_rp_mean_mag'][i],ngc['parallax'][i]) for i in range(len(ngc['ra']))]
M_b = [abs_mag(ngc['phot_bp_mean_mag'][i],ngc['parallax'][i]) for i in range(len(ngc['ra']))]
color_index = color(M_rp,M_b)
plt.figure(figsize=(7,7))
plt.title('Color-Magnitude - Initial Data')
plt.xlabel('(Bp-Rp)')
plt.ylabel('Absolute Magnitude ')
plt.xlim([0,2.5])
plt.ylim([31,22])
plt.scatter(color_index, M_g,marker='o', s=25,alpha = .5, c = 'k')
plt.style.use('seaborn')
plt.show() 
```
